```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "ipl_matches_2008_2022.csv"
df = pd.read_csv(file_path)

df["match_date"] = pd.to_datetime(df["match_date"], errors="coerce")

# 1. Matches per Season
plt.figure(figsize=(12, 6))
df["season"].value_counts().sort_index().plot(kind='bar', color='blue')
plt.title("Number of Matches per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Top Winning Teams
plt.figure(figsize=(12, 6))
df["winning_team"].value_counts().plot(kind='barh', color='orange')
plt.title("Top Winning Teams")
plt.xlabel("Number of Wins")
plt.ylabel("Team")
plt.show()
```

```python
# 3. Toss Decision Analysis
plt.figure(figsize=(6, 6))
df["toss_decision"].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)
plt.title("Toss Decisions: Bat vs. Field")
plt.ylabel("")
plt.show()
```

```python
# 4. Winning Margin Distribution
plt.figure(figsize=(10, 6))
plt.hist(df["margin"].dropna(), bins=30, color='teal', edgecolor='black', alpha=0.7)
plt.title("Winning Margin Distribution")
plt.xlabel("Winning Margin (Runs/Wickets)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. Venue Popularity
plt.figure(figsize=(12, 6))
df["venue"].value_counts().nlargest(10).plot(kind='barh', color='purple')
plt.title("Top 10 Venues by Matches Played")
plt.xlabel("Number of Matches")
plt.ylabel("Venue")
plt.show()
```

```python
# 6. Winning Methods Analysis
plt.figure(figsize=(6, 6))
df["won_by"].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'orange'], startangle=90)
plt.title("Winning Methods: Runs vs. Wickets")
plt.ylabel("")
plt.show()
```

```python
# 7. Top Players of the Match Awards
plt.figure(figsize=(12, 6))
df["player_of_match"].value_counts().nlargest(10).plot(kind='bar', color='red')
plt.title("Top 10 Players with Most Player of the Match Awards")
plt.xlabel("Player")
plt.ylabel("Number of Awards")
plt.xticks(rotation=45)
plt.show()
```

```python
# 8. Most Frequent Finals Teams
finals_df = df[df["match_number"] == "Final"]
plt.figure(figsize=(12, 6))
finals_df["team1"].value_counts().add(finals_df["team2"].value_counts(), fill_value=0).sort_values(ascending=False).plot(kind='bar', color='green')
plt.title("Most Frequent Finals Teams")
plt.xlabel("Team")
plt.ylabel("Finals Appearances")
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. Toss Winners vs. Match Winners
plt.figure(figsize=(6, 6))
(df["toss_winner"] == df["winning_team"]).value_counts().plot(kind='pie', labels=["Same", "Different"], autopct='%1.1f%%', colors=['green', 'red'], startangle=90)
plt.title("Toss Winner Also Winning the Match")
plt.ylabel("")
plt.show()
```

```python
# 10. Matches Per City
plt.figure(figsize=(12, 6))
df["city"].value_counts().nlargest(10).plot(kind='bar', color='blue')
plt.title("Top 10 Cities Hosting IPL Matches")
plt.xlabel("City")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)
plt.show()
```
